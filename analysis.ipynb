```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import pandas as pd
import numpy as np

npart = 5
steps = 100
# Replace 'your_file.csv' with the actual path to your CSV file
file_path = 'data.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

#Sample data
x = np.array([df['x'].iloc[i:i+npart].to_numpy() for i in np.arange(steps)])
print(np.size(x[i]))
y = np.array([df['y'].iloc[i:i+npart].to_numpy() for i in np.arange(steps)])
z = np.array([df['z'].iloc[i:i+npart].to_numpy() for i in np.arange(steps)])

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
i = 90
ax.scatter(x[1], y[1], z[1], c='r', marker='o')
ax.scatter(x[i], y[i], z[i], c='g', marker='o')
ax.scatter(np.linspace(x[0], x[i], 10), np.linspace(y[0], y[i], 10), np.linspace(z[0], z[i], 10), c='b', s=1)

# Set axis labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Show the plot
plt.show()
```
